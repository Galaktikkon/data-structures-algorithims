Problem maksymalnego przepywu

1. Wejcie:
 - $G=(V,E)$ - graf skierowany (dla kadych $u$,$v \in V$, co najwyej jedna z $(u,v)$, $(v,u)$ istnieje)
 - $s$,$t \in V$ - rdo i ujcie
    - do $t$ tylko krawdzie wchodzce
    - z $s$ tylko krawdzie wychodzce
 - $c: V\times V \rightarrow \N$ - funkcja pojemnoci:
    - $c(u,v)=0 \Leftrightarrow (u,v) \notin E$ - Pojemno krawdzi wynosi $0$, wtedy i tylko wtedy, gdy ona nie istnieje
    - $c(u,u)=0$ - w grafie nie ma ptli

Zadanie: znale przepyw o maksymalnej wartoci 

2. Przepyw - jest to funkcja:

$f: V\times V \rightarrow \N$ - $f$ spenia nastpujce warunki:
 - $(\forall u,v \in V)[f(u,v) \leq c(u,v)]$ - wpyw do wierzchoka moe by mniejszy lub rwny jego pojemnoci
 - $(\forall v \in V -\{s,t\})[\sum_{u \in V} f(u,v) = \sum_{u \in V} f(v,u)]$ - tyle ile wpywa do wierzchoka, tyle wypywa

Warto przepywu: 
 - $|f|=\sum_{u \in V} f(s,v) - \sum_{u \in V} f(v,s)$ - suma wypyww z $s$ pomniejszona o sum wpywajcych ($=0$)





3. Sieci residualne:
 - $G=(V,E)$
 - $s,t \in V$
 - $c: V \times V \rightarrow \N$
 - $f: V \times V \rightarrow \N$

Sie przepywowa $+$ przepyw $=$ sie residualna

Definiujemy sie residualn $G_f (G_f = G)$, $c_f$ przez funkcj $c_f: V\times V \rightarrow \N$:
 - $(u,v) \in E \Rightarrow c(u,v) - f(u,v)$ - jeeli istnieje przepyw, to pomniejszamy pojemno o to co wypywno
 - $(v,u) \in E \Rightarrow f(v,u)$
 - w pp.: $0$




4. cieka powikszajca dla $G,f,c$, to cieka z $s$ do $t$ w sieci residualnej ($G_f, c_f$)

Wartoci cieki powikszajcej jest minimalna warto $c_f(u,v)$ na ciece:
 - wzdu cieki powikszajcej moemy powikszy przepyw o warto tej cieki

5. Metoda Forda-Fulkersona:
 - jeli istnieje cieka powikszajca dla $G,f$, to powiksz $f$ wzdlu cieki
 - wr do kroku 1

6. Przekrj w sieci:
 - $G=(V,E)$
 - $s,t \in V$
 - $c: V \times V \rightarrow \N$
 - $f: V \times V \rightarrow \N$
 - Przekrj sieci, to podzia zbioru $V$ na dwa zbiory:
   - $S, T=V-S$, gdzie $ s \in S$, $t \in T$
 - Przepustowo przekroju to:
   - $c(S,T) = \sum_{u \in S} \sum_{v \in T} c(u,v)$
 - Przepyw netto przez przekrj:
   - $f(S,T) = (\sum_{u \in S} \sum_{v \in T} f(u,v)) - (\sum_{u \in S} \sum_{v \in T} f(u,v))$



$Lemat$: </br>
Niech $f$ bdzie przepywem w sieci $G$ ze rdem $s$, ujciem $t$, gdzie $(S,T)$, jest przekrojem. Wwczas $f(S,T)=|f|$

$Dowd$: </br>
z warunku zachowania przepywu:
 - $(\forall v \in V -\{s,t\})[\sum_{u \in V} f(u,v) - \sum_{u \in V} f(v,u)] = 0$
 - $|f|=\sum_{u \in V} f(s,v) - \sum_{u \in V} f(v,s) + \sum_{u \in S-\{s\}} \sum_{v \in V} f(u,v) - \sum_{u \in S-\{s\}} \sum_{v \in V} f(v,u)$ </br> $=  \sum_{u \in S} \sum_{v \in V} f(u,v) -\sum_{u \in S} \sum_{v \in V} f(v,u)$ </br> $= \sum_{u \in S} \sum_{v \in T} f(u,v) -\sum_{u \in S} \sum_{v \in T} f(v,u)$ </br> $+ \sum_{u \in S} \sum_{v \in S} f(u,v) -\sum_{u \in S} \sum_{v \in S} f(v,u)$ </br>$= f(S,T) + 0$ </br> $= f(S,T)$

$Lemat$: </br>
Warto przepywu jest niewiksza ni przepustowo dowolnego przekroju

$Dowd$: </br>
$|f| = f(S,T)$ </br> $= \sum_{u \in S} \sum_{v \in S} f(u,v) - \sum_{u \in S} \sum_{v \in S} f(v,u)$ </br> $\leq \sum_{u \in S} \sum_{v \in S} f(u,v)$ </br> $\leq \sum_{u \in S} \sum_{v \in T} c(u,v) = c(S,T) $ (z $def$ przepustowoci przekroju)

7. Twierdzenie o maksymalnym przepywie i minimalnym przekroju (max-flow/min-cut theorem):

Niech $G=(V,E)$,$s,t \in V$, $c: V \times V \rightarrow \N$ - bdzie sieci przepywow, a $f$ bdzie przepywem w tej sieci.
Nastpujce warunki s rwnowane:
 - $f$ jest maksymalnym przepywem (1)
 - $G_f, c_f$ nie zawiera cieek powikszajcych (2)
 - dla pewnego przekroju $S,T$ zachodzi: (3)
   - $|f| = c(S,T)$

$Dowd$:
 - (3) $\Rightarrow$ (1) - natychmiastowe spostrzerzenie z poprzedniego lematu
 - (1) $\Rightarrow$ (2) - oczywiste, bo inaczej moglibymy powikszy przepyw i nie byby maksymalny

Chcemy pokaza, e (2) $\Rightarrow$ (3), bo wtedy domkniemy ptle dowodzenia:
 - $S=\{ v \in V \}$: istnieje cieka z $s$ do $v$ w $G_f$:
   - $s \in S, t \notin S, T = V-S, t \in T$
 - $f(S,T) = \sum_{u \in S} \sum_{v \in T}(f(u,v) - f(v,u)) $
 - Jeli $(u,v) \in E$, to $c_f(u,v) = 0 \Rightarrow f(u,v) = c(u,v)$
 - Jeli $(v,u) \in E$, to $c_f(u,v) = 0 \Rightarrow f(v,u) = 0$
 - $= \sum_{u \in S} \sum_{v \in T} c(u,v) = c(S,T) $ 

8. Zoono czasowa metody Forda-Fulkersona: $O((V+E)|f^{*}|) = O(E|f^{*}|)$, gdzie:
 - $V+E$ - zoonoc szuakania cieek powikszajcych
 - $|f^{*}|$ - warto maksymalnego przeplywu
 - Jeli zawsze ciek powikszajc znajdujemy algorytmem BFS, to zoono wynosi $O(VE^2)$

9. Znajdowanie najlicznejszego skojarzenia w grafach dwudzielnych:
 - Graf dwudzielny, to graf, ktrego zbir wierzchokw mona podzieli na czsci $U$ i $V$, takie e kada krawd zawiera jeden wierzchoek z $U$ i jeden z $V$.
 - Skojarzeniem nazywamy, zbir krawdzi, takich e kade dwie nie maj wsplnego wierzchoka.



Najliczniejsze skojarzenie znajdujemy przez obliczenie maksymalnego przepywu:
- $O(V^3)$
