1. Wejście:
 - $G=(V,E)$ - graf skierowany (dla każdych $u$,$v \in V$, co najwyżej jedna z $(u,v)$, $(v,u)$ istnieje)
 - $s$,$t \in V$ - źródło i ujście
    - do $t$ tylko krawędzie wchodzące
    - z $s$ tylko krawędzie wychodzące
 - $c: V\times V \rightarrow \N$ - funkcja pojemności:
    - $c(u,v)=0 \Leftrightarrow (u,v) \notin E$ - Pojemność krawędzi wynosi $0$, wtedy i tylko wtedy, gdy ona nie istnieje
    - $c(u,u)=0$ - w grafie nie ma pętli

Zadanie: znaleźć przepływ o maksymalnej wartości 

2. Przepływ - jest to funkcja:

$f: V\times V \rightarrow \N$ - $f$ spełnia następujące warunki:
 - $(\forall u,v \in V)[f(u,v) \leq c(u,v)]$ - wpływ do wierzchołka może być mniejszy lub równy jego pojemności
 - $(\forall v \in V -\{s,t\})[\sum_{u \in V} f(u,v) = \sum_{u \in V} f(v,u)]$ - tyle ile wpływa do wierzchołka, tyle wypływa

Wartość przepływu: 
 - $|f|=\sum_{u \in V} f(s,v) - \sum_{u \in V} f(v,s)$ - suma wypływów z $s$ pomniejszona o sumę wpływających ($=0$)





3. Sieci residualne:
 - $G=(V,E)$
 - $s,t \in V$
 - $c: V \times V \rightarrow \N$
 - $f: V \times V \rightarrow \N$

Sieć przepływowa $+$ przepływ $=$ sieć residualna

Definiujemy sieć residualną $G_f (G_f = G)$, $c_f$ przez funkcję $c_f: V\times V \rightarrow \N$:
 - $(u,v) \in E \Rightarrow c(u,v) - f(u,v)$ - jeżeli istnieje przepływ, to pomniejszamy pojemność o to co wypływnęło
 - $(v,u) \in E \Rightarrow f(v,u)$
 - w pp.: $0$




4. Ścieżka powiększająca dla $G,f,c$, to ścieżka z $s$ do $t$ w sieci residualnej ($G_f, c_f$)

Wartością ścieżki powiększającej jest minimalna wartość $c_f(u,v)$ na ścieżce:
 - wzdłuż ścieżki powiększającej możemy powiększyć przepływ o wartość tej ścieżki

5. Metoda Forda-Fulkersona:
 - jeśli istnieje ścieżka powiększająca dla $G,f$, to powiększ $f$ wzdluż ścieżki
 - wróć do kroku 1

6. Przekrój w sieci:
 - $G=(V,E)$
 - $s,t \in V$
 - $c: V \times V \rightarrow \N$
 - $f: V \times V \rightarrow \N$
 - Przekrój sieci, to podział zbioru $V$ na dwa zbiory:
   - $S, T=V-S$, gdzie $ s \in S$, $t \in T$
 - Przepustowość przekroju to:
   - $c(S,T) = \sum_{u \in S} \sum_{v \in T} c(u,v)$
 - Przepływ netto przez przekrój:
   - $f(S,T) = (\sum_{u \in S} \sum_{v \in T} f(u,v)) - (\sum_{u \in S} \sum_{v \in T} f(u,v))$



$Lemat$: </br>
Niech $f$ będzie przepływem w sieci $G$ ze źródłem $s$, ujściem $t$, gdzie $(S,T)$, jest przekrojem. Wówczas $f(S,T)=|f|$

$Dowód$: </br>
z warunku zachowania przepływu:
 - $(\forall v \in V -\{s,t\})[\sum_{u \in V} f(u,v) - \sum_{u \in V} f(v,u)] = 0$
 - $|f|=\sum_{u \in V} f(s,v) - \sum_{u \in V} f(v,s) + \sum_{u \in S-\{s\}} \sum_{v \in V} f(u,v) - \sum_{u \in S-\{s\}} \sum_{v \in V} f(v,u)$ </br> $=  \sum_{u \in S} \sum_{v \in V} f(u,v) -\sum_{u \in S} \sum_{v \in V} f(v,u)$ </br> $= \sum_{u \in S} \sum_{v \in T} f(u,v) -\sum_{u \in S} \sum_{v \in T} f(v,u)$ </br> $+ \sum_{u \in S} \sum_{v \in S} f(u,v) -\sum_{u \in S} \sum_{v \in S} f(v,u)$ </br>$= f(S,T) + 0$ </br> $= f(S,T)$

$Lemat$: </br>
Wartość przepływu jest niewiększa niż przepustowość dowolnego przekroju

$Dowód$: </br>
$|f| = f(S,T)$ </br> $= \sum_{u \in S} \sum_{v \in S} f(u,v) - \sum_{u \in S} \sum_{v \in S} f(v,u)$ </br> $\leq \sum_{u \in S} \sum_{v \in S} f(u,v)$ </br> $\leq \sum_{u \in S} \sum_{v \in T} c(u,v) = c(S,T) $ (z $def$ przepustowości przekroju)

7. Twierdzenie o maksymalnym przepływie i minimalnym przekroju (max-flow/min-cut theorem):

Niech $G=(V,E)$,$s,t \in V$, $c: V \times V \rightarrow \N$ - będzie siecią przepływową, a $f$ będzie przepływem w tej sieci.
Następujące warunki są równoważne:
 - $f$ jest maksymalnym przepływem (1)
 - $G_f, c_f$ nie zawiera ścieżek powiększających (2)
 - dla pewnego przekroju $S,T$ zachodzi: (3)
   - $|f| = c(S,T)$

$Dowód$:
 - (3) $\Rightarrow$ (1) - natychmiastowe spostrzerzenie z poprzedniego lematu
 - (1) $\Rightarrow$ (2) - oczywiste, bo inaczej moglibyśmy powiększyć przepływ i nie byłby maksymalny

Chcemy pokazać, że (2) $\Rightarrow$ (3), bo wtedy domkniemy pętle dowodzenia:
 - $S=\{ v \in V \}$: istnieje ścieżka z $s$ do $v$ w $G_f$:
   - $s \in S, t \notin S, T = V-S, t \in T$
 - $f(S,T) = \sum_{u \in S} \sum_{v \in T}(f(u,v) - f(v,u)) $
 - Jeśli $(u,v) \in E$, to $c_f(u,v) = 0 \Rightarrow f(u,v) = c(u,v)$
 - Jeśli $(v,u) \in E$, to $c_f(u,v) = 0 \Rightarrow f(v,u) = 0$
 - $= \sum_{u \in S} \sum_{v \in T} c(u,v) = c(S,T) $ 

8. Złożoność czasowa metody Forda-Fulkersona: $O((V+E)|f^{*}|) = O(E|f^{*}|)$, gdzie:
 - $V+E$ - złożonośc szuakania ścieżek powiększających
 - $|f^{*}|$ - wartość maksymalnego przeplywu
 - Jeśli zawsze ścieżkę powiększającą znajdujemy algorytmem BFS, to złożoność wynosi $O(VE^2)$


Implementacja dla grafu w postaci macierzowej:

```python
def bfs(G,a,b):
    n=len(G)
    s=a
    t=b
    from collections import deque
    visited=[False for _ in range(n)]
    visited[s]=True
    parent=[None for _ in range(n)]
    path=[]
    Q=deque()
    Q.append(a)
    while Q:
        u=Q.popleft()
        if u==t:
            while t!=None:
                path.append(t)
                t=parent[t]
            return path[::-1]
        for i in range(n):
            if not visited[i] and G[u][i]:
                visited[i]=True
                parent[i]=u
                Q.append(i)
                
    return path[::-1]

def cap(G,path):
    s=G[path[0]][path[1]]
    for i in range(1,len(path)-1):
        s=min(s,G[path[i]][path[i+1]])
    return s

def update(G,path):
    w=cap(G,path)
    for i in range(len(path)-1):
        G[path[i]][path[i+1]]-=w
        G[path[i+1]][path[i]]+=w

def FordFulkerson(G,s,t):
    from copy import deepcopy
    flow=0
    F=deepcopy(G)
    path=bfs(F,s,t)
    while path:
        flow+=cap(F,path)
        update(F,path)
        path=bfs(F,s,t)
    return flow

E1 = [(0, 1, 10),(1, 2, 25), (2, 3, 10), (0, 5, 10), (5, 4, 15), (4, 1, 6), (4, 3, 10)]

E2 = [(0, 3, 10), (3, 1, 20), (3, 6, 15), (1, 2, 10), (2, 5, 15), (4, 3, 3), (4, 1, 15), (5, 4, 4),
     (5, 8, 10), (7, 5, 7), (7, 4, 10), (6, 7, 10), (9, 0, 10), (9, 1, 5), (9, 2, 10), (6, 10, 15),
     (8, 10, 10)]

def get_graph(E):
    n=-1
    for e in E:
        n=max(n,e[1],e[0])
    n+=1
    G=[[0 for _ in range(n)] for _ in range(n)]
    
    for e in E:
        G[e[0]][e[1]]=e[2]
        
    return G

def ford_fulkerson_check(G,s,t):
    A=get_graph(G)
    print("maximum flow:",FordFulkerson(A,s,t))

ford_fulkerson_check(E1,0,3)
ford_fulkerson_check(E2,9,10)
```


9. Znajdowanie najlicznejszego skojarzenia w grafach dwudzielnych:
 - Graf dwudzielny, to graf, którego zbiór wierzchołków można podzielić na częsci $U$ i $V$, takie że każda krawędź zawiera jeden wierzchołek z $U$ i jeden z $V$.
 - Skojarzeniem nazywamy, zbiór krawędzi, takich że każde dwie nie mają wspólnego wierzchołka.

Najliczniejsze skojarzenie znajdujemy przez obliczenie maksymalnego przepływu:
- $O(V^3)$
